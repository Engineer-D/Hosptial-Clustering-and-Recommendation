#### Categorical encoding


Our model can only work with numerical type features so we need to convert all object type, bool type and date type into numerical type


for columns having just 2/3 unique type we will make use of one hot encoding

for columns with more unique type we make use of frequency encoding

##### Frequency Encoding

```python
def freq_encode(data,column_index, drop=False):
  '''function to frequency encode a column in the data'''
  col=str(data.columns[column_index])
  count=data.groupby(col).size()
  frequency=count/len(data)
  data[col+'frequency']=data[col].map(frequency)
  if drop==True:
    data.drop(col,axis=1,inplace=True)
  
  return data
```

```python
dfcc = pd.read_csv('Health_Dataset.csv')
gps=dfcc[['X_gps_latitude','X_gps_longitude']]
```

```python
#add gps location
dfcc=pd.concat([dfc,gps], axis=1)
```

```python
def clustering(data, zone=None, state=None, lga=None, grid_proximity=None):
    '''Function to cluster the Health centers based on state, zone.
    
       data: Health dataset
       zone: any of the options: south_east, north_central, south_south, north_west,
       north_east, south_west.
       lga: local government area
       state: state
       grid_proximity: nearness of the health center. Options: connected, beyond_500m, within_500m, dk
    '''

    if zone==None or state==None:
        raise ValueError("Invalid request: Its compulsory to provide zone and state info. lga and grid_proximity is optional.")
    elif zone !=None and state!=None and lga!=None:
        zone,state,lga=zone.lower(), state.lower(), lga.lower()
        result1= data[(data.zone==zone) & (data.state==state) & (data.lga==lga)]['facility_type'].values
        result2=data[(data.zone==zone) & (data.state==state) & (data.lga==lga)]['grid_proximity'].values
        result3=data[(data.zone==zone) & (data.state==state) & (data.lga==lga)]['X_gps_latitude'].values
        result4=data[(data.zone==zone) & (data.state==state) & (data.lga==lga)]['X_gps_longitude'].values
        final_result=pd.DataFrame()
        final_result['Hospital']=result1
        final_result['grid _proximity']=result2
        final_result['latitude']=result3
        final_result['longitude']=result4
        if grid_proximity != None:
            final_result=final_result[final_result['grid _proximity']==grid_proximity]
            if len(final_result)==0:
                raise ValueError('sorry no hospital with such proximity')
    elif zone !=None and state!=None:
        zone,state=zone.lower(), state.lower()
        result1= data[(data.zone==zone) & (data.state==state)]['facility_type'].values
        result2= data[(data.zone==zone) & (data.state==state)]['lga'].values
        result3=data[(data.zone==zone) & (data.state==state)]['grid_proximity'].values
        result4=data[(data.zone==zone) & (data.state==state)]['X_gps_latitude'].values
        result5=data[(data.zone==zone) & (data.state==state)]['X_gps_longitude'].values
        final_result=pd.DataFrame()
        final_result['Hospital']=result1
        final_result['lga']=result2
        final_result['grid _proximity']=result3
        final_result['latitude']=result4
        final_result['longitude']=result5
        if grid_proximity != None:
            final_result=final_result[final_result['grid _proximity']==grid_proximity]
            if len(final_result)==0:
                raise ValueError('sorry no hospital with such proximity')
    return final_result
                
```
